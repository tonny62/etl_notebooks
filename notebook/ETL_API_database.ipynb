````python
def seeKeyValue(key, dataList):
    ''' See the value of the key'''
    result = []
    for item in dataList:
        if(item.get(key)):
            result.append(item[key])
    return result
````

````python
import pandas as pd
import mysql.connector

def getConnection():
    return mysql.connector.connect(user='jobdatabaseroot', password='jobdatabaseroot', host='jobdatabase2.cieq9tyvhuue.ap-southeast-1.rds.amazonaws.com', database='job_database')

def readIntoDF(query, arguments=None):
    try :
        cnx = getConnection()
        if(arguments != None):
            if (type(arguments) != tuple):
                raise TypeError("arguments is not of type tuple")
            cursor = cnx.cursor()
            cursor.execute(query, arguments)
        else:
            cursor = cnx.cursor()
            cursor.execute(query)

        header = [item if type(item) == str else item.decode() for item in cursor.column_names]
        outData = pd.DataFrame(data=cursor.fetchall(), columns=header)
    finally:
        cnx.close()
    return outData

def update(query, arguments=None):
    try:
        if(arguments != None):
            if (type(arguments) != tuple):
                raise TypeError("arguments is not of type tuple")
            cnx = getConnection()
            cursor = cnx.cursor()
            cursor.execute(query, arguments)
        else:
            cnx = getConnection()
            cursor = cnx.cursor()
            cursor.execute(query)
        cnx.commit()
        return True
    finally:
        cnx.close()
````

````python
import random
## create jobtitle
def getjobtitle(jobdict):
    key = ['func', 'pos', 'pos2', 'posth']
    title = ''
    for item in key:
        if(jobdict.get(item)!=None):
            title = title + " " + jobdict.get(item)
    return title

## create ostarnetid
def getostarnetid(jobdict):
    ## wait from classifier
    return random.randint(1,4)

## create companyid
def getcompanyid(jobdict):
    query = "SELECT companyid FROM company WHERE companyname = %s;"
    company = jobdict.get('company') if jobdict.get('company') != None else jobdict.get('com')
    result = readIntoDF(query, (company,))
    if(result.shape[0] == 0):
        ## could not find in database
            ## get industry id first then insert
            update("INSERT INTO company(`companyname`, `provinceid`, `industryid`, `jpno`) VALUES(%s,%s,%s,%s)", (company, getprovinceid(jobdict), random.randint(1,4), random.randint(0,100000)))
            return getcompanyid(jobdict)
    else:
        ## found in the database
        return result['companyid'][0]
        
    
## create amount 1
def getamount(jobdict):
    return random.randint(1,5)

## create postdate
def getpostdate(jobdict):
    ## MySQL use format YYYY-MM-DD
    ## input is 12-May-2017
    monthdict = {"January":1, "February":2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9, "October":10, "November":11, "December":12}
    date = jobdict.get('pdate').split("-")
    date = "20"+date[2]+"-"+"0"*(2-len(str(monthdict.get(date[1]))))+str(monthdict.get(date[1]))+"-"+date[0]
    return date

## create edulevel list 2
def getedulevel(jobdict):
    return [random.randint(1,5) for item in range(0,random.randint(1,2))]

## create major list 3
def getmajor(jobdict):
    return [random.randint(1,8) for item in range(0,random.randint(1,3))]

## create keyword list 
def getkeyword(jobdict):
    return [random.randint(0,item) for item in range(0,random.randint(1,10))]

from thai import *
## create provinceid
def getProvinceIndex(jobdict):
    ''' Return the PK of the province
    0. find เขต -> BKK
    1. find [0-9]{5} for postal address map to province
    2. find province -> province ID
    3. find district -> provinceID '''
    try:
        location_column = ['district', 'location', 'loc_det', 'loc', 'com_loc']
        location_str = " ".join([word for word in [jobdict.get(item) for item in location_column] if word != None])
    except:
        return 78
    
    if(len(location_str) == 0):
        return 78

    ## 0. find 'เขต'
    if('เขต' in location_str):
        ## Bangkok
        return 1
    
    ## 1. try to find postal code
    import re
    from checklang import isTH
    postal_re = re.compile(r"[0-9]{5}")
    if(postal_re.search(location_str) != None):
        provinceCode = getProvinceCode_postal(postal_re.search(location_str)[0])
        if(provinceCode != None):
            return provinceCode
    
    ## 2. map province to province
    provinceCode = getProvinceCode_prov(location_str)
    if(provinceCode != None):
        return provinceCode
        
    ## 3. find province inside the string
    provinceCode = getProvinceCode_inProv(location_str)
    if(provinceCode != None):
        return provinceCode

    ## 4. map district to province
    provinceCode = getProvinceCode_district(location_str)
    if(provinceCode != None):
        return provinceCode
    
    ## 5. laven dist
    if(len(location_str) <= 25):
        provinceCode = getProvinceCode_minEdit(location_str)
        return provinceCode
    
    return None

def getprovinceid(jobdict):
    result = getProvinceIndex(jobdict)
    if(result!=None):
        return result
    else:
        return 78
````

````python
## Transform into jobid, jobtitle, ostarnetid, companyid, amount, postdate, edulevel:list, major:list, keyword:list, provinceid
import pandas as pd
import time

start = time.time()

data_in = jobs[0:5]
data_out = pd.DataFrame(columns=['jobid','jobtitle','ostarnetid', 'companyid', 'amount', 'postdate', 'edulevel', 'major', 'keyword', 'provinceid'])

## create dataframe index
data_out['jobid'] = range(0, len(data_in))

## call extraction methods
data_out['jobtitle'] = [getjobtitle(item) for item in data_in]
data_out['ostarnetid'] = [getostarnetid(item) for item in data_in]
data_out['companyid'] = [getcompanyid(item) for item in data_in]
data_out['amount'] = [getamount(item) for item in data_in]
data_out['postdate'] = [getpostdate(item) for item in data_in]
data_out['edulevel'] = [getedulevel(item) for item in data_in]
data_out['major'] = [getmajor(item) for item in data_in]
data_out['keyword'] = [getkeyword(item) for item in data_in]
data_out['provinceid'] = [getprovinceid(item) for item in data_in]

end = time.time()
print("time taken")
print(end - start)
````

```

3. get major from text

4. province (above)
```

````python
def loadProcedure(data):
    data_out = data
    try:
        cnx = getConnection()
        cursor = cnx.cursor()

        for index, series in data_out.iterrows():
            # insert job
            query = '''
                INSERT INTO job(`jobtitle`, `ostarnetid`, `postdate`, `amount`, `companyid`, `provinceid`) VALUES (%s, %s, %s, %s, %s ,%s);
            '''
            data = (series['jobtitle'], series['ostarnetid'], series['postdate'], series['amount'], series['companyid'], series['provinceid'])
            cursor.execute(query, data)
            lastrowid = cursor.lastrowid

            query = '''
                INSERT INTO job_has_edulevel(`jobid`, `edulevelid`)
                VALUES(%s, %s);
            '''
            data = tuple((lastrowid, item) for item in series['edulevel'])
            cursor.executemany(query, data)

            query = '''
                INSERT INTO job_has_major(`jobid`, `majorid`)
                VALUES(%s, %s);
            '''
            data = tuple((lastrowid, item) for item in series['major'])
            cursor.executemany(query, data)

    #         query = '''
    #             INSERT INTO job_has_keyword(`jobid`, `keywordid`)
    #             VALUES(%s, %s);
    #         '''
    #         data = tuple((lastrowid, item) for item in set(series['keyword']))
    #         cursor.executemany(query, data)

            cnx.commit()
    finally :
        cursor.close()
        cnx.close()
````
